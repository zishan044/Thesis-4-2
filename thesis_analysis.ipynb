```python
import random

def convert_age(age):
  return int(age[:2])+random.randint(0,2)

def convert_gender(gender):
  return int(gender == 'Male')

def convert_personality(personality):
  personality_map = {'Introvert': 0, 'Ambivert': 1, 'Extrovert': 2}
  return personality_map[personality]

def convert_preference(preference):
  
  preference_map = {
      "Light mode always": 0,
      "Dark mode for certain apps, light mode for others": 2,
      "Dark mode in low light, otherwise light mode": 2,
      "Dark mode rarely, light mode usally": 1,
      "Dark mode always": 3
  }
  return preference_map[preference]

def convert_dark_default(dark_default):
  return int(dark_default == 'Yes')

def convert_color_choice(color_choice):
  return int(color_choice)-1

def convert_hours(hour):
  return min(9, int(hour[:1])+random.randint(0,2))

ages = np.array([convert_age(age) for age in df['age']])
genders = np.array([convert_gender(gender) for gender in df['gender']])
personalities = np.array([convert_personality(personality) for personality in df['personality']])
preferences = np.array([convert_preference(preference) for preference in df['dark_mode_preference']])
dark_defaults = np.array([convert_dark_default(dark_default) for dark_default in df['default_dark_mode']])
color_choices = np.array([convert_color_choice(color_choice) for color_choice in df['color_choice']])
hours = np.array([convert_hours(hour) for hour in df['hours_on_devices']])

def make_scatter_plot(x, y, x_label, y_label):
  plt.figure(figsize=(10, 6))
  plt.scatter(x, y, s=500, alpha=0.04)
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.show()

make_scatter_plot(ages, preferences, x_label="Age", y_label="Dark Mode Preference")
make_scatter_plot(genders, preferences, x_label="Gender", y_label="Dark Mode Preference")
make_scatter_plot(personalities, preferences, x_label="Personality", y_label="Dark Mode Preference")

make_scatter_plot(ages, dark_defaults, x_label="Age", y_label="Dark Mode Default")
make_scatter_plot(genders, dark_defaults, x_label="Gender", y_label="Dark Mode Default")
make_scatter_plot(personalities, dark_defaults, x_label="Personality", y_label="Dark Mode Default")
```

```python
from scipy.stats import chi2_contingency

def chi_stat(x, y):
  table = np.zeros(shape=(len(set(x)), len(set(y))))
  min_x, min_y = x.min(), y.min()

  for i in range(len(x)):
    table[x[i]-min_x][y[i]-min_y] += 1

  return chi2_contingency(table)


chi, p, dof, exp = chi_stat(ages, preferences)
print(chi, p)
chi, p, dof, exp = chi_stat(genders, preferences)
print(chi, p)
chi, p, dof, exp = chi_stat(personalities, preferences)
print(chi, p)

chi, p, dof, exp = chi_stat(ages, color_choices)
print(chi, p)
chi, p, dof, exp = chi_stat(genders, color_choices)
print(chi, p)
chi, p, dof, exp = chi_stat(personalities, color_choices)
print(chi, p)

chi, p, dof, exp = chi_stat(ages, dark_defaults)
print(chi, p)
chi, p, dof, exp = chi_stat(genders, dark_defaults)
print(chi, p)
chi, p, dof, exp = chi_stat(personalities, dark_defaults)
print(chi, p)


chi, p, dof, exp = chi_stat(hours, preferences)
print(chi, p)  

# table = np.zeros(shape=(len(set(ages)), len(set(preferences))))

# for i in range(len(ages)):
#   table[ages[i]-18][preferences[i]] += 1

# print(chi2_contingency(table))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from scipy.stats import chisquare

df = pd.read_csv('data.csv')

age = np.array([int(x) + random.randint(0,2) for x in df['age']])
color = np.array([int(x) for x in df['color']])
hours = np.array([int(x) + random.randint(0,1) for x in df['hours']])
google = np.array([int(x) for x in df['google']])
tech = np.array([int(x) for x in df['tech']])
familiar = np.array([int(x) for x in df['familiar']])
preference = np.array([int(x) for x in df['preference']])

def summary(arr, label):

  mean = np.mean(arr)
  median = np.median(arr)
  std = np.std(arr)

  x, y = np.unique(arr, return_counts=True)
  mode = np.argmax(y)

  print(f"Summary statistics for {label}")
  print(f"Mean: {mean:.2f}")
  print(f"Median: {median:.2f}")
  print(f"Mode: {mode:.2f}")
  print(f"Standard Deviation: {std:.2f}")

arrs = [age, color, hours, google, tech, familiar]
labels = ['Age', 'Color choice', 'Hours spent on devices', 'Googling tendency', 'Tech enthusiasm', 'Familiarity with dark mode', 'Preference for using dark mode']

for i in range(6):
  summary(arrs[i], labels[i])
```

```python
from scipy.stats import chi2_contingency

hours = np.array([int(x) + random.randint(0,1) for x in df['hours']])
preference = np.array([int(x) for x in df['preference']])

plt.scatter(hours, preference, s=500, alpha=0.04)
plt.title("How device use time relates to dark mode\n", fontsize=20)
plt.xlabel('Device use hours\n', fontsize=15)
plt.ylabel("Preference of dark mode\n", fontsize=15)

t = pd.crosstab(hours, preference, rownames=['hours'], colnames=['preference'])
print(t)
def chi_stat(x, y):
  table = np.zeros(shape=(len(set(x)), len(set(y))))
  min_x, min_y = x.min(), y.min()

  for i in range(len(x)):
    table[x[i]-min_x][y[i]-min_y] += 1

  return chi2_contingency(table)


chi, p, dof, exp = chi_stat(hours, preference)
print(f"Chi Square Stat = {chi:.4},\nP Value = {p:.2f}")
corr = np.corrcoef(hours, preference)[0][1]
print(f"Correlation Coefficient: {corr:.2f}")
```
